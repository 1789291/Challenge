```python
import os
import json
import torch
import cv2
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from pathlib import Path

# Set the paths for the images and annotations
IMAGES_PATH = "coin-dataset/"
ANNOTATIONS_PATH = "coin-dataset/_annotations.coco.json"
MODEL_WEIGHTS_PATH = "./yolov5s.pt"  # Replace with your model weights

# Load the trained YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'custom', path=MODEL_WEIGHTS_PATH)
```

```python
# Load annotations from JSON file
def load_annotations(json_path):
    with open(json_path, 'r') as f:
        annotations = json.load(f)
    return annotations
```

```python

# Function to detect coins in an image
def detect_coins(image_path, model):
    # Load the image with OpenCV or PIL
    img = Image.open(image_path)

    # Run inference using the YOLOv5 model
    results = model(img)

    # Extract predictions
    predictions = results.xyxy[0]  # x1, y1, x2, y2, confidence, class
    return predictions

```

```python
# Function to visualize the detections
def visualize_detections(image_path, detections):
    img = cv2.imread(image_path)
    for detection in detections:
        x1, y1, x2, y2, conf, cls = detection
        
        # Draw bounding boxes
        cv2.rectangle(img, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)

        # Add class and confidence label
        label = f'Coin {conf:.2f}'
        cv2.putText(img, label, (int(x1), int(y1)-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Convert image to RGB for Matplotlib
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Display the image with bounding boxes
    plt.figure(figsize=(10, 8))
    plt.imshow(img_rgb)
    plt.axis('off')
    plt.show()
```

```python
# Main function to process images and detect coins
def main():
    # Load the annotations
    annotations = load_annotations(ANNOTATIONS_PATH)

    # Process each image in the dataset
    for image_data in annotations['images']:
        image_name = image_data['file_name']
        image_path = os.path.join(IMAGES_PATH, image_name)

        
        print(f"Processing {image_name}...")

        # Detect coins in the image
        detections = detect_coins(image_path, model)
        
        # Visualize the detection results
        visualize_detections(image_path, detections)

if __name__ == "__main__":
    main()
```
